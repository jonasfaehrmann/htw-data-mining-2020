```python
import pandas as pd
from currency_converter import CurrencyConverter
```

# Descriptive Analysis

```python
data = pd.read_csv('./data/test.csv')
data
```

```python
# Compute pound values to eur (currency_in_eur)
c = CurrencyConverter()
def convert_to_eur(row):
    if row['Currency'] == '£':
        return c.convert(row['Price'], 'EUR', 'GBP')
    return row['Price']

data['currency_in_eur'] = data.apply(convert_to_eur, axis=1)
```

```python
# Analysis of currency_in_eur column
data['currency_in_eur'].describe()
```

```python
# Analysis of currency_in_eur column per route_abb
print(data.groupby(['route_abb']).mean())
print(data.groupby(['route_abb']).std())
```

```python
# Analysis of currency_in_eur column per date
print(data.groupby(['Flight_Date']).mean())
print(data.groupby(['Flight_Date']).std())
```
